```python
import selenium
import pandas as pd
import time
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
driver.get("https://www.indiacensus.net/literacy-rate.php")
```

```python
population=[]
try:
    pop_list=driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[4]/div[2]/table/tbody[1]/tr/td[3]")
    for i in pop_list:
        population.append(i.text)
except NoSuchElementException:
        population.append("-") 
population
```

```python
len(population)
```

```python
population = [int(s.replace(',', '')) for s in population]

print(population)
```

```python
state_ut=[]
try:
    s_u_list=driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[4]/div[2]/table/tbody[1]/tr/td[2]")
    for i in s_u_list:
        state_ut.append(i.text)
except NoSuchElementException:
        state_ut.append("-") 
state_ut
```

```python
len(state_ut)
```

```python
literacy=[]
try:
    literacy_list=driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[4]/div[2]/table/tbody[1]/tr/td[5]")
    for i in literacy_list:
        literacy.append(i.text)
except NoSuchElementException:
        literacy.append("-") 
literacy
```
