# Question no.: 8
Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
from any YouTube Video. 

```python
driver=webdriver.Chrome()       
driver.get("https://www.youtube.com/watch?v=PlPgYOFJROI")
time.sleep(2)
```

```python
search_bar=driver.find_element(By.XPATH,"//input[@id='search']").send_keys("DIY craft")
search=driver.find_element(By.XPATH,"//*[@id='search-icon-legacy']").click()
time.sleep(2)
```

```python
vedio=driver.find_element(By.XPATH,"//*[@id='video-title']/yt-formatted-string")
vedio.click()
```

```python
for _ in range(1000):
    driver.execute_script("window.scrollBy(0,1000)")
```

```python
comments=[]
time_update=[]
vote_up=[]
comment=driver.find_elements(By.XPATH,"//div[@class='style-scope ytd-expander']")
for i in comment:
    comments.append(i.text)

time=driver.find_elements(By.XPATH,"//a[@class='yt-simple-endpoint style-scope yt-formatted-string']")
for i in time:
    time_update.append(i.text)
    
vote=driver.find_elements(By.XPATH,"//span[@class='style-scope ytd-comment-action-buttons-renderer']")
for i in vote:
    vote_up.append(i.text)
```
